使用上一个月的销售数量作为答案，是一个可以最为基准线的分数

```python
# coding: utf-8
import os

import pandas as pd
import numpy as np
import scipy

% matplotlib inline
```

```python
sales_df = pd.read_csv('./data/input/sales_train.csv')
tst_df = pd.read_csv('./data/input/test.csv')
```

```python
sales_df['revenue'] = sales_df['item_price'] * sales_df['item_cnt_day']
trn_df = sales_df.groupby(['shop_id', 'item_id', 'date_block_num'])['item_cnt_day'].sum().reset_index()
trn_df = trn_df.rename(columns={'item_cnt_day': 'item_cnt_month'})
```

```python
trn_df[trn_df['date_block_num'] == 33]
```

```python
tst_df = tst_df.merge(trn_df[trn_df['date_block_num'] == 33],
                      on=['shop_id', 'item_id'], how='left'
                     ).fillna(0)
tst_df['item_cnt_month'] = tst_df['item_cnt_month'].clip(0, 20)
tst_df
```

```python
tst_df['ID'] = tst_df['ID'].astype(int).values
tst_df[['ID', 'item_cnt_month']].to_csv('./data/output/submission/previous_month_benchmark.csv', index=False)
```
